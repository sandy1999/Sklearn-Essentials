```python
# importing numpy for matrix maths 
import numpy as np 

# for plotting 
import matplotlib.pyplot as plt 

from scipy import stats

# for beautyful graphs 
import seaborn as sns; sns.set()

%matplotlib inline
```

# Motivating Suport Vector Machines 

We have learned about simple classification line underlying the curve and creating a line or distribution but here we are going to use a more simplified model to classify the datapoints.

Lets perform a simple classification task:

```python
# making a data set 
from sklearn.datasets.samples_generator import make_blobs

# making a sample data of two kinds 
X,y = make_blobs(n_samples= 50, centers= 2 , random_state=0, cluster_std=0.60)

# plotting the graph 
plt.scatter(X[:,0],X[:,1], c=y, s=50, cmap='autumn')
```

A linear discriminative classifier can draw a line seperating two datasets for the classification process and we can draw this line by hand but there is this problem i.e. multiple lines can be drawn and we need to select which one is the best one.

Lines which can be drawn by these plots:

```python
# making a xfit data 
xfit = np.linspace(-1, 3.5)

# scatter plot 
plt.scatter(X[:,0],X[:,1], c=y, s=50, cmap='autumn')

# plotting a mark point 
plt.plot([0.6],[2.1],'X',color='red', markersize=10, markeredgewidth = 2)

# plotting lines of different slope

for m,b in [(1,0.65),(0.5,1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')
    
# setting plot limit 
plt.xlim(-1, 3.5);
```

There are different of ways in which we can seperate these class but consider a point with completely new data point *X* of new type so we instead of just fitting the line we need to think about the bigger picture.

# SVM: Maximizing the Margin

Support vectors machines provides one way to this. The intution is instead of drawing a zero-width line segment we can can draw a line segement of enough width *margin* which can pass nearby to near points of line segment, this may look like:

```python
# plotting the marginal line 
# generating xfit data 
xfit = np.linspace(-1, 3.5)

# plotting the scatter 
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='autumn')

# plotting lines
for m,b,d in [(1.0, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit,yfit - d, yfit + d, edgecolor = 'none', color='#AAAAAA', alpha = '0.4') # to make margin
    
plt.xlim(-1, 3.5);
```

We choose the line with maximum boundary in the support vector machines or the ones with the *maximum margin* is the optimal model for the SVM's. 

## Fitting a support vector machine

Let's see the result of actual data fit in the model of SVM classifier of *Scikit-learn*. We are going to use kernel as linear and set high value of `c` parameter.

```python
# importing SCV 
from sklearn.svm import SVC

# making a model 
model = SVC(kernel='linear', C=1E10)

# fitting the model data
model.fit(X,y)
```

To visualize what is happening over here lets create a quick convininece function to that will plot decision boundaires to us:

```python
def plot_svc_decision_function(model, ax = None, plot_support = True):
    """Plot the decision function for 2D SVM"""
    
    # defining the axes of plot
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create a grid to evaluate a model 
    x = np.linspace(xlim[0],xlim[1], 30)
    y = np.linspace(ylim[0],ylim[1], 30)
    
    Y,X = np.meshgrid(y,x)
    
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k', levels=[-1,0,1], linestyles = ['--','-','--'], alpha = 0.5)
    
    # plot support vector
    if plot_support:
        ax.scatter(model.support_vectors_[:,0],
                  model.support_vectors_[:,1],
                  s=300, linewidth = 1, facecolors = 'none')
        
    # plot boundaries
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
# plotting the actual plot 
plt.scatter(X[:,0], X[:,1], c = y, s=50, cmap='autumn')
plot_svc_decision_function(model)
```

This is the dividing line that maximizes the margin between the two sets of the points. Notice that a few of the training points just the touch training points which are also the pivotial points known as *support vectors*, and give this algorithm the name it needs and they are stored in a array of `support_vectors_` of *Scikit-Learn*.

```python
# support vectors for above fit
model.support_vectors_
```

A key to this classifier's fit is the nearby points to margin and *Support Vectors* which are responsible for the fit, far away points don't contribute to the line and if a new data point is added it also doesnt affect if it is placed and classsified correctly.


We can see that for example if we plot the model for first 60 points, and first 120 points:\

```python
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

In the left pannel we can see the model and support vectors for 60 training points and in right pannel we have doubled the training points still support vectors remains the same that is the biggest strength of **SVM's**, there insensitivity towards the fit, makes them the one of the most used algorithm.

## Beyond linear boundraies: Kernel SVM

The true power of SVM's are unleashed when there are combined with *Kernels*.Kernels are the functions that are used to convert our data into higher dimensions in order to fit non lineraly seperable data.

In SVM models, we can use a version of the same idea. To motivate the need for kernels lets look at the example:

```python
# making a circular dataset 
from sklearn.datasets.samples_generator import make_circles

# a circular datasets
X,y = make_circles(n_samples=100, noise=0.1, factor=0.1)

# fitting the model 
clf = SVC(kernel='linear').fit(X,y)

plt.scatter(X[:,0],X[:,1], c=y, cmap='autumn',s=50)
plot_svc_decision_function(clf,plot_support=False)
```

It is clear that no linear discriminative will be able to seperate this model so we are need of something better and more robhust model which can be used to make a seperation and can act as an medicore to classify this class, here is the time when kernel tricks comes to the play. 

We can think of to project these datapoints into a higher dimesion, where linear seperator would be sufficient for the case, we cam use *radical basis function* centered at middle clump:

```python
# radical values 
r = np.exp((X ** 2).sum(1))
r[:5]
```

We can visualize the extra dimesion using 3d plot and if you are using this notebook live then you can also interact with the plot.

```python
# making a 3d plot 

from mpl_toolkits import mplot3d

# making a plot function 
def plot_3D(elev=30, azim=30, X=X, y=y):
    ax = plt.subplot(projection = '3d')
    ax.scatter3D(X[:,0], X[:,1], c=y, s=50, cmap='autumn')
    ax.view_init(elev=elev, azim=azim)
    
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')
    
interact(plot_3D, elev = [-90, 90], azim = (-180, 180), X = fixed(X), y = fixed(y))
```

Here we have to choose carefully to choose our transformation which can be done with the help of kernel tricks, we can say that above data becomes linearly seperable after transforming it to radial basis.

So to fit our data we can use `rbf` kernel in `SVC` class for our transformation.

```python
# creating a clf 
clf = SVC(kernel='rbf', C=1e6)
clf.fit(X,y)
```

```python
# plotting the clf 
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], facecolors='none', s=300, lw = 1);
```

Using this kernelized support vector machine, we learn a suitable nonlinear decision boundary. This transformation technique is often used in machine learning to turn fast linear to fast non linear methods where *kernel tricks* can be used.

## Tuning the SVM:Softening Margins

Our discusion have so far centered around the clean dataset in which there was no overlapping and perfect decision boundary exsist. But what if our data set has some extent of overlapping? For example:

```python
# making a dataset 
X,y = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=1.2)

# plotting the scatter 
plt.scatter(X[:,0],X[:,1],c=y,s=50, cmap='autumn');
```

To implement this SVM has many fudge factors to soften the margin that is allow some of the data points to creep into margin boundary if it allows a better fit. This property is controlled by a paramenter $C$. Higher the value of $C$ harder is the margin that is it doesnt allows the points to creep into margin and lower is the value of $C$ *softer* the margin parameter and can grow to encompass the datapoints.

Here is an example defining how a change in parameter $C$ changes the plot:

```python
# making datasets 
X,y = make_blobs(n_samples=100, centers=2, cluster_std=0.80, random_state=0)

# creating a subplot
fig,ax = plt.subplots(1,2,figsize = (16 , 6))
fig.subplots_adjust(left = 0.0625, right = 0.95, wspace = 0.1)

# plotting for various C 
for axi, c in zip(ax,[10.0, 0.1]):
    clf = SVC(kernel='linear',C=c).fit(X,y) # fitting a model 
    
    axi.scatter(X[:,0],X[:,1], c=y, s=50, cmap='autumn')
    
    plot_svc_decision_function(clf,axi)
    axi.scatter(clf.support_vectors_[:,0],
                clf.support_vectors_[:,1],
                s=300, lw = 1, facecolors='none');
    axi.set_title('C = {0:.1f}'.format(c),size=14)
```

The optimal value of our $C$ parameter will depend on dataset and should be tuned accordingly using cross-validation or similar product(for more details refer [Hyperparameters and Model Validation](./Hyperparameters and Model Validation.ipynb))
